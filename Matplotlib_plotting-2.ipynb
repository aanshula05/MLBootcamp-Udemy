```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
ds = pd.read_csv('Heart_Disease_Prediction.csv')
ds
```

```python
ds['Heart Disease'] = ds['Heart Disease'].map({'Presence': 1, 'Absence': 0})
```

```python
ds
```

```python
ds.plot(x='BP', y='Max HR');
```

```python
ds['Cholesterol'].plot.hist(color='red');
```

```python
ds.plot(kind='scatter', x='Cholesterol', y='BP');
```

```python
ds.plot.hist(subplots=True, figsize=(15,20));
```

```python
over_50 = ds[ds['Age'].astype(int)> 50]
over_50
```

```python
ds.shape
```

```python
len(over_50)
```

```python
over_50.plot(kind='scatter', x='Age', y='Cholesterol', c='Heart Disease');
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='Age', y='Cholesterol', c='Heart Disease', ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='Age', y='Cholesterol', c='Heart Disease', ax=ax);

#to adjust axis limits
ax.set_xlim([45,100]);
```

```python
fig, ax = plt.subplots(figsize=(10,6))

#to create scatter plot
scatter = ax.scatter(x=over_50['Age'], y=over_50['Cholesterol'], c=over_50['Heart Disease'])

#customize scatter plot
ax.set(title="Heart Disease Prediction", xlabel='Age', ylabel='Cholesterol');

#to add legend
ax.legend(*scatter.legend_elements(), title="Heart Disease");

#to add a horizontal line
ax.axhline(over_50['Cholesterol'].mean(), linestyle='dotted', color="red")
```

```python
over_50.head()
```

```python
#subplots of max HR, age, cholesterol
fig, (ax0,ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True) #sharex since both have the same x axis

#ax0
scatter = ax0.scatter(x = over_50['Age'], y=over_50['Cholesterol'], c=over_50['Heart Disease'])
ax0.set(title="Age Vs Cholesterol", ylabel='Cholesterol');
ax0.legend(*scatter.legend_elements());
ax0.axhline(over_50['Cholesterol'].mean());


#ax1
scatter = ax1.scatter(x=over_50['Age'], y=over_50['Max HR'], c=over_50['Heart Disease'])
ax1.set(title='Age Vs Max HR', xlabel='Age', ylabel='Max HR');
ax1.legend(*scatter.legend_elements());
ax1.axhline(over_50['Max HR'].mean(), linestyle='dotted');

#add a title to the figure
fig.suptitle('Heart Disease Prediction', fontsize=16, fontweight="bold");
```

```python
plt.style.use('seaborn-whitegrid');
fig, ax = plt.subplots(figsize=(10,6))

#to create scatter plot
scatter = ax.scatter(x=over_50['Age'], y=over_50['Cholesterol'], c=over_50['Heart Disease'], 
                    cmap='summer') #this changes colour scheme

#customize scatter plot
ax.set(title="Heart Disease Prediction", xlabel='Age', ylabel='Cholesterol');

#to add legend
ax.legend(*scatter.legend_elements(), title="Heart Disease");

#to add a horizontal line
ax.axhline(over_50['Cholesterol'].mean(), linestyle='dotted', color="red")
```

```python
plt.style.use('seaborn-whitegrid');
fig, ax = plt.subplots(figsize=(10,6))

#to create scatter plot
scatter = ax.scatter(x=over_50['Age'], y=over_50['Cholesterol'], c=over_50['Heart Disease'], 
                    cmap='viridis') #this changes colour scheme

#customize scatter plot
ax.set(title="Heart Disease Prediction", xlabel='Age', ylabel='Cholesterol');

#to add legend
ax.legend(*scatter.legend_elements(), title="Heart Disease");

#to add a horizontal line
ax.axhline(over_50['Cholesterol'].mean(), linestyle='dotted', color="red")
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x = over_50['Age'], y=over_50['Cholesterol'], c=over_50['Heart Disease'], cmap='winter');
ax.set(title='Heart Disease Prediction')
ax.legend(*scatter.legend_elements());
ax.axhline(over_50['Cholesterol'].mean(), color='orange', linestyle='--');
```

```python
#customizing x and y limits
#subplots of max HR, age, cholesterol
fig, (ax0,ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True) #sharex since both have the same x axis

#ax0
scatter = ax0.scatter(x = over_50['Age'], y=over_50['Cholesterol'], c=over_50['Heart Disease'], cmap='winter')
ax0.set(title="Age Vs Cholesterol", ylabel='Cholesterol');
ax0.legend(*scatter.legend_elements());
ax0.axhline(over_50['Cholesterol'].mean());
ax0.set_xlim([50,80])

#ax1
scatter = ax1.scatter(x=over_50['Age'], y=over_50['Max HR'], c=over_50['Heart Disease'], cmap='summer')
ax1.set(title='Age Vs Max HR', xlabel='Age', ylabel='Max HR');
ax1.legend(*scatter.legend_elements());
ax1.axhline(over_50['Max HR'].mean(), linestyle='dotted');
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#add a title to the figure
fig.suptitle('Heart Disease Prediction', fontsize=16, fontweight="bold");
```
