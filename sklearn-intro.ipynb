# An end-to-end scikit-learn workflow

```python
import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv('Heart_Disease_Prediction.csv')
heart_disease
```

```python
heart_disease['Heart Disease'].map({'Presence': 0, 'Absence': 1})
```

```python
heart_disease['Heart Disease'] = heart_disease['Heart Disease'].map({'Presence': 0, 'Absence': 1})
```

```python
heart_disease.head()
```

```python
heart_disease = heart_disease.rename(columns = {'Heart Disease': 'Target'})
heart_disease
```

```python
#Create X(features matrix)
X = heart_disease.drop('Target', axis=1)

#create y (labels)
y = heart_disease['Target']
```

```python
#to know version of sklearn
import sklearn
sklearn.show_versions()
```

```python
#using random forest classifier
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()  #clf->classifier

#clf = RandomForestClassifier(n-estimators = 100) it is 100 by default
#keep default hyperparameters    
clf.get_params()
```

```python
#fit the model to data(splitting into training and test data)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf.fit(X_train,y_train)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
y_label = clf.predict(np.array([1,2,3,4]))
```

```python
#to make predictions 
y_pred = clf.predict(X_test)
y_pred
```

```python
#evaluate the model on training data and test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test) 
```

Hence accuracy of 79%

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred)
```

```python
#to improve the model
#1. try different amts of n-estimators
np.random.seed(0)

for i in range(10, 110, 10):
    print(f"Trying model with {i} estimators")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set {clf.score(X_test, y_test) *100}")
    print("")
```

```python
#save a model and load it
import pickle
pickle.dump(clf, open('random_forest1.pkl', "wb"))  #wb-> write binary
```

```python
loaded_model = pickle.load(open("random_forest1.pkl", "rb"))  #rb-> read binary
loaded_model.score(X_test, y_test)
```
