```python
Grocery_list_prices = {"Milk": 25,
                       "Bread": 50,
                       "Butter": 75}
```

```python
#to plot horizontal bars
fig, ax = plt.subplots()
ax.barh(list(Grocery_list_prices.keys()), list(Grocery_list_prices.values()));
```

# Subplots
Two options for subplots

```python
#Option1 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
#Plot to each different axis
ax1.plot(x, x**2);
ax2.plot(x, x/2);
ax3.scatter(np.random.random(20), np.random.random(20));
ax4.hist(y);
```

```python
x = np.linspace(0,10,100)
x
```

```python
#option 2
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,5))

#plotting
ax[0].plot(x, x**2);
ax[1].plot(x,np.sin(x));
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

#plotting
ax[0,0].plot(x,x/2);
ax[0,1].plot(x,np.tan(x));
ax[1,0].scatter(np.random.random(20), np.random.random(20));
ax[1,1].hist(y);
```
