```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#linscape returns evenly spaced nos in given interval
x = np.linspace(0,10,100)
x
```

```python
#create line plot
fig , ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
```

```python
#create scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, x**4);
```

```python
fig, ax =plt.subplots()
ax.scatter(x, np.sin(x), color="orange");
```

```python
#creating a plot directly from a dictionary
grocery_item_prices = {"Milk": 25,
                       "Bread": 50,
                       "Butter": 75}
fig, ax = plt.subplots()
ax.bar(grocery_item_prices.keys(),grocery_item_prices.values(), color="green");
ax.set(title="Grocery Item Prices", ylabel="Price($)");
```
