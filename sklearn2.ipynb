To convert columns Make and Colour to numerical data

```python
X = car_sales.drop('Price', axis=1)
y = car_sales['Price']
```

```python
#Split into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Building machine learning model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

ML model can't deal with strings

```python
#to conv into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_features = ['Make', 'Colour', 'Doors']

one_hot = OneHotEncoder() #one hot encoding is the process of conv categories to nos
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)],remainder='passthrough')

transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
pd.DataFrame(transformed_X)
```

```python
#converting using pandas
dummies = pd.get_dummies(car_sales[['Make', 'Colour', 'Doors']])
```

```python
dummies
```

```python
#refit the model
np.random.seed(0)
X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
