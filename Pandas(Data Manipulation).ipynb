# Manipulating data

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Colour"].str.lower()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]','').astype(int)
```

```python
car_sales
```

```python
car_sales["Colour"] = car_sales["Colour"].str.lower()
```

```python
car_sales
```

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
#Filling missing data
car_sales_missing["Odometer"].mean()
```

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=True)
```

```python
car_sales_missing
```

```python
#Remove missing values
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
car_sales_missing_dropped.to_csv("car_sales_missing_dropped")
```

```python
#Column from series
seats_column = pd.Series([5,5,5,5,5])  

#Column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5,inplace=True)
car_sales
```

```python
#Column from list
fuel_economy = [7.5, 9.6, 3.2, 4.5, 6.6, 7.3, 8.1, 9.4, 4.9, 5.1]
car_sales["fuel per 100km"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["fuel per 100km"]
car_sales
```

```python
#Column from a single value
car_sales["No of wheels"] = 4
car_sales
```

```python
car_sales["Passed road safely"] = True
car_sales.dtypes
```

```python
car_sales
```

```python
#to remove a column
car_sales = car_sales.drop("No of wheels", axis = 1)
car_sales
```

```python
#to shuffle rows
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
#to only select 30% of data
car_sales_shuffled.sample(frac=0.3)
```

```python
#to get back to original data
car_sales_shuffled.reset_index(drop=True, inplace=True)
car_sales_shuffled
```

```python
#to convert km to miles
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x/1.6)
car_sales
```
