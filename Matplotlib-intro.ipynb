# Matplotib Workflow

```python
#1. Import libraries
%matplotlib inline
import matplotlib.pyplot as plt

#2. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#3. Setup Plot
fig, ax = plt.subplots(figsize=(5,5))

#4. Plot the data
ax.plot(x,y)

#5. Customize plot
ax.set(title="Simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

#6. Save and show figure
fig.savefig("Images/sample-plot.png")

```
